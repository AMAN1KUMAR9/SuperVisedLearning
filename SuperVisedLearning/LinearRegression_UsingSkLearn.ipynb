```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

data = pd.read_csv("data_for_lr.csv")
data = data.dropna()
```

```python
train_input = np.array(data.x[0:500] ).reshape(500 , 1)
train_output = np.array(data.y[0:500] ).reshape(500 , 1)
test_input = np.array(data.x[500:700] ).reshape(199 , 1)
test_output = np.array(data.y[500:700] ).reshape(199 , 1)
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression

lr_model = LinearRegression() # initialize model
lr_model.fit(train_input , train_output)
lr_model.coef_ 
lr_model.intercept_
```

```python
test_predictions = lr_model.predict(test_input)

plt.plot(test_input , test_predictions , "*" , color  = "green")
plt.plot(test_input , test_output , "*" , color  = "yellow")
plt.title("model testing")
plt.xlabel("Input")
plt.ylabel("Ouput/Predictions")
plt.show()
```

### cost

```python
from sklearn.metrics import mean_squared_error

mean_squared_error(test_output,test_predictions)
```
